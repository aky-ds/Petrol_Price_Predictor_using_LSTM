## Petrol Prices are going up continously these days and it has become very hard to predict what will be the price range in the next few months or even days.

### In this project we will try to forecast the different petrol prices for the upcoming dates from the dataset has been given to us.  

### We will use LSTM and ARIMA forecasting models to predict the prices then we will see how we can implement the same using Auto Keras, which is an Auto ML Library.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('train_data.csv')
```

```python
df.head()
```

```python
df.shape
```

<h1> Data Preprocessing and Cleaning </h1>

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.duplicated().sum()
```

```python
sns.displot(x=df['Petrol (USD)'])
```

```python
sns.boxplot(x=df['Petrol (USD)'])
```

```python
df3=df[df['Petrol (USD)']<160]
plt.scatter(df3.Date, df3['Petrol (USD)'],color='blue',label='Prices', s=50)
```

```python
df=df3.reset_index()['Petrol (USD)']
```

<h2> Making the Features </h2>

```python
# For LSTM We need Scaling 
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler(feature_range=(0,1))
df=scaler.fit_transform(np.array(df).reshape(-1,1))
```

```python
training_size=int(len(df)*(0.65))
testing_size=len(df)-training_size
```

```python
train_data,test_data=df[0:training_size,:],df[training_size:len(df),:1]
```

```python
train_data.shape
```

```python
test_data.shape
```

```python
def create_dataset(dataset, time_step=1):
    dataX, dataY = [], []
    for i in range(len(dataset)-time_step-1):
        a = dataset[i:(i+time_step), 0]   ###i=0, 0,1,2,3-----99   100 
        dataX.append(a)
        dataY.append(dataset[i + time_step, 0])
    return np.array(dataX), np.array(dataY)  # Arrays function have been created
```

```python
# Applying the train_test_split
time_step=100
X_train,y_train=create_dataset(train_data,time_step)
X_test,y_test=create_dataset(test_data,time_step)
```

```python
X_train.shape
```

```python
X_train=X_train.reshape(X_train.shape[0],X_train.shape[1], 1)
X_test=X_test.reshape(X_test.shape[0],X_test.shape[1], 1)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,LSTM
```

```python
X_train.shape
```

```python
# Creating an RNN network
```

```python
from tensorflow.keras.models import Sequential

from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LSTM
model=Sequential()
model.add(LSTM(100,return_sequences=True,input_shape=(100,1)))
model.add(LSTM(80,return_sequences=True))
model.add(LSTM(70))
model.add(Dense(1))
model.compile(loss='mean_squared_error',optimizer='adam')
```

```python
history=model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=100,batch_size=64,verbose=1)
```

```python
### Lets Do the prediction and check performance metrics
train_predict=model.predict(X_train)
test_predict=model.predict(X_test)
```

```python
##Transformback to original form
train_predict=scaler.inverse_transform(train_predict)
test_predict=scaler.inverse_transform(test_predict)
```

```python
### Plotting 
# shift train predictions for plotting
look_back=100
trainPredictPlot = np.empty_like(df)
trainPredictPlot[:, :] = np.nan
trainPredictPlot[look_back:len(train_predict)+look_back, :] = train_predict
# shift test predictions for plotting
testPredictPlot = np.empty_like(df)
testPredictPlot[:, :] = np.nan
testPredictPlot[len(train_predict)+(look_back*2)+1:len(df)-1, :] = test_predict
# plot baseline and predictionss
plt.plot(scaler.inverse_transform(df),label='actual_data')
plt.plot(trainPredictPlot,label='train_line')
plt.plot(testPredictPlot,label='test_line')
plt.legend()
plt.show()
```

```python
model.save('Petrol Price Predictor.h5')
```
